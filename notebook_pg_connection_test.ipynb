```python
query = """select 1 as number;"""
pd.read_sql(query, con=engine)

query = """
    select * from pg_catalog.pg_tables
    where schemaname != 'pg_catalog' and schemaname != 'information_schema';
"""
pd.read_sql(query, con=engine)

```

```python
from time import time
import pandas as pd

df = pd.read_csv('popular_anime.csv', nrows=10)
df_iter = pd.read_csv('popular_anime.csv', iterator=True, chunksize=100000)
df.head(0).to_sql(name='popular_anime', con=engine, if_exists='replace', index=False)

t_start = time()

count = 0
for batch in df_iter:
    count+=1
    batch_df = batch 

    print(f'inserting batch {count}...')

    b_start = time()
    batch_df.to_sql(name='popular_anime', con=engine, if_exists='append', index=False )
    b_end = time() 

    print(f'inserted! time talem {b_end-b_start:10.3f} seconds. \n')

t_end = time()
print(

    f'Completed! Total time taken was {t_end-t_start:10.3f} seconds for {count} batches.')
```
